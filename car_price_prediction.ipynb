```python
a = pd.get_dummies(df['make'])
b = pd.get_dummies(df['fuel-type'])
c = pd.get_dummies(df['aspiration'])
d = pd.get_dummies(df['num-of-doors'])
e = pd.get_dummies(df['body-style'])
f = pd.get_dummies(df['drive-wheels'])
g = pd.get_dummies(df['engine-location'])
h = pd.get_dummies(df['engine-type'])
i = pd.get_dummies(df['num-of-cylinders'])
j = pd.get_dummies(df['fuel-system'])

```

```python
df = pd.concat([df, a], axis=1)
df = pd.concat([df, b], axis=1)
df = pd.concat([df, c], axis=1)
df = pd.concat([df, d], axis=1)
df = pd.concat([df, e], axis=1)
df = pd.concat([df, f], axis=1)
df = pd.concat([df, g], axis=1)
df = pd.concat([df, h], axis=1)
df = pd.concat([df, i], axis=1)
df = pd.concat([df, j], axis=1)

```
